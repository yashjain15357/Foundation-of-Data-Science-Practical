### Develop program for Frequency distributions


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

import data from csv

```python
# Load the dataset
student_data = pd.read_csv(r'C:\Users\YASH JAIN\Documents\New Folder\OneDrive\Desktop\CodeBox\Foundation-of-Data-Science-Practical\students.csv')

# Rename the DataFrame variable
student_df = student_data
```

Histogram

```python
# Plot a histogram for the 'Age' column with a custom color
plt.figure(figsize=(8, 5))
plt.hist(student_df['Age'], bins=8, color='skyblue', edgecolor='black')
plt.title('Distribution of Student Ages', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Bar Plot

```python
# Create a bar plot for 'City' vs 'Fee'
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Fee', data=student_df, palette='viridis')
plt.title('Average Fee by City', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Fee', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

Box Plot

```python
plt.figure(figsize=(10, 5))  # Set the figure size
plt.boxplot(student_df["Age"], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue", color="blue"))
q1 = np.percentile(student_df['Age'], 25)  # First quartile (Q1)
q2 = np.percentile(student_df['Age'], 50)  # Median (Q2)
q3 = np.percentile(student_df['Age'], 75)  # Third quartile (Q3)

# Add annotations for Q1, Q2, and Q3
plt.text(q1, 1.05, f"Q1: {q1:.2f}", ha="center", fontsize=10, color="darkblue")
plt.text(q2, 1.1, f"Q2 (Median): {q2:.2f}", ha="center", fontsize=10, color="darkgreen")
plt.text(q3, 1.05, f"Q3: {q3:.2f}", ha="center", fontsize=10, color="darkred")

# Add labels and title
plt.xlabel("Age", fontsize=12)
plt.title("Box Plot of Student Ages with Q1, Q2, Q3", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

Frequency Polygon

```python
# Frequency polygon with updated data
age = [12, 15, 18, 20, 22, 25, 28, 30, 35, 40, 42, 45, 50, 55, 60, 65, 70, 75, 80, 85]
interval = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# Set custom ticks for the x-axis
plt.xticks(interval)

# Create a histogram with step type
y, edges, _ = plt.hist(age, bins=interval, histtype='step', edgecolor='k', linewidth=1.5)

# Calculate midpoints of intervals
midpoint = 0.5 * (edges[1:] + edges[:-1])

# Plot the frequency polygon
plt.plot(midpoint, y, marker='o', color='blue', linestyle='-', linewidth=2)

# Add grid, labels, and title
plt.grid(alpha=0.5, linestyle='--')
plt.xlabel('Age Intervals', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Frequency Polygon of Ages', fontsize=14)
plt.show()
```
