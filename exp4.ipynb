```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

import data from csv

```python
students = pd.read_csv('https://raw.githubusercontent.com/arz03/fds-lab/refs/heads/main/students.csv') # write students.csv in file
df = students # dataframe variable
#df.head() # no need to write
```

Histogram

```python
plt.hist(df['Fee'],bins=10)
plt.xlabel("Fees")
plt.ylabel("Frequency")
plt.title("Frequency Distribution Histogram")
plt.show()
```

Bar Plot

```python
df['City'].value_counts().plot(
    kind='bar',
    xlabel='City', ylabel='Frequency',
    title='Frequency Distribution Bar Plot')
```

Box Plot

```python
plt.boxplot(df["Age"], vert=False)
q1 = np.percentile(df['Age'], 25)  # First quartile (Q1)
q2 = np.percentile(df['Age'], 50)  # Median (Q2)
q3 = np.percentile(df['Age'], 75)  # Third quartile (Q3)
plt.text(q1, 1.05, f"Q1: {q1}", ha="center")
plt.text(q2, 1.1, f"Q2 (Median): {q2}", ha="center")
plt.text(q3, 1.05, f"Q3: {q3}", ha="center")
plt.xlabel("Age")
plt.title("Box Plot of Student Ages with Q1, Q2, Q3")
plt.grid(axis="x")
plt.figure(figsize=(8, 5)) # no need to write
```

```python
ages = df["Age"]

# Compute quartiles and interquartile range (IQR)
q1 = np.percentile(ages, 25)  # First quartile (Q1)
q2 = np.percentile(ages, 50)  # Median (Q2)
q3 = np.percentile(ages, 75)  # Third quartile (Q3)
iqr = q3 - q1  # Interquartile Range

# Compute outliers (1.5*IQR rule)
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = [age for age in ages if age < lower_bound or age > upper_bound]

# Create a box plot
plt.figure(figsize=(8, 5))
box = plt.boxplot(ages, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))

# Annotate Q1, Q2, Q3
plt.text(q1, 1.05, f"Q1: {q1}", ha="center", fontsize=12, color="blue")
plt.text(q2, 1.1, f"Q2 (Median): {q2}", ha="center", fontsize=12, color="red")
plt.text(q3, 1.05, f"Q3: {q3}", ha="center", fontsize=12, color="blue")

# Annotate outliers (if any)
for outlier in outliers:
    plt.text(outlier, 1.15, f"{outlier}", ha="center", fontsize=12, color="black")

    # Labels and title
    plt.xlabel("Age")
    plt.title("Box Plot of Student Ages with Q1, Q2, Q3, and Outliers")
    plt.grid(axis="x", linestyle="--", alpha=0.7)

    # Show the plot
    plt.show()
```
