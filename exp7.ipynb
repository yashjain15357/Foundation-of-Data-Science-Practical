### Develop program for Normal Curves

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Updated dataset of student scores
scores = [55, 60, 62, 65, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100, 102, 105, 108, 110]

# Calculate mean and standard deviation
mean_score = np.mean(scores)
std_dev_score = np.std(scores)
print(f'Mean = {mean_score:.2f}')
print(f'Standard Deviation = {std_dev_score:.2f}')

# Generate a range of x-values for the normal distribution
x_values = np.linspace(mean_score - 3 * std_dev_score, mean_score + 3 * std_dev_score, 100)

# Calculate the probability density function (PDF) for the normal distribution
pdf_values = stats.norm.pdf(x_values, mean_score, std_dev_score)

# Create the plot
plt.plot(x_values, pdf_values, label='Normal Distribution', color='blue', linewidth=2)

# Add title, labels, and legend
plt.title('Normal Distribution of Student Scores', fontsize=14)
plt.xlabel('Scores', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```
